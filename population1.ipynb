## 연령별 인구 시각화

```python
import csv
import matplotlib.pyplot as plt

f = open('./datas/age_2023.csv', 'r')
data = csv.reader(f)
next(data)
result = []  # 연령별 인구 저장할 리스트

for row in data:
    if '강남구 논현2동' in row[0]:
        for i in row[3:]:  # 3번 인덱스 ~ 103번 인덱스까지 반복
            result.append(int(i.replace(",", '')))  # 천단위 구분기호 replace로 제거
            
print(len(row[3:]))
#print(result)
plt.rc('font', family='Malgun Gothic')
plt.style.use("ggplot")  # 격자 무늬 스타일
plt.title("연령별 인구 통계")
plt.plot(result, 'r') # result - y축값
plt.xlabel('연령')
plt.ylabel('인구')
plt.show()
```
